#### Write a Python function to determine whether a given string is a palindrome.
Requirements:
    Ignore spaces, punctuation, and case sensitivity.
    Return True if the string is a palindrome, otherwise False.

```python
def is_palindrome(strr:str) -> bool:
    strr = "".join(i.lower() for i in strr if i.isalnum())
    return strr == strr[::-1]

print(is_palindrome("madam"))                 # True
print(is_palindrome("RaceCar"))               # True
print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome("hello"))                 # False
```

#### Write a function to find the first non-repeating character in a string.
Return the character itself. If no such character exists, return None.

```python
def first_non_repeating(s:str):
    counter = {}
    for ch in s:
        counter[ch] = counter.get(ch,0)+1
    for ch in s:
        if counter[ch]==1:
            return ch
    return None

print(first_non_repeating('leetcode'))
print(first_non_repeating('aabbcc'))
print(first_non_repeating('aabbccd'))
print(first_non_repeating('swiss'))
```

#### Write a function to remove duplicate characters from a string while preserving the original order.

```python
def remove_duplicates(s:str)-> str:
    seen = set()
    result = []
    
    for ch in s:
        if ch not in seen:
            seen.add(ch)
            result.append(ch)
            
    return "".join(result)

print(remove_duplicates('programming'))
print(remove_duplicates('aabbcc'))
print(remove_duplicates('hello'))
print(remove_duplicates(''))
```

#### Write a function to count the number of vowels in a string.

```python
def count_vowels(s:str) -> int:
    counter = 0
    for ch in s.lower():
        if ch in {'a','e', 'i', 'o', 'u'}:
            counter +=1
    return counter

print(count_vowels('Hello'))
print(count_vowels('PYTHON'))
print(count_vowels('aeiou'))
print(count_vowels('bcdfg'))
```

#### Write a function to find the second largest number in a list.
Requirements: Do not use sort() or sorted(). If there is no second largest number (e.g., [5] or [5, 5, 5]), return None.

```python
def second_largest(numbers:list) -> int:
    if len(numbers)<2:
        return None
    
    largest = None
    second = None
    for num in numbers:
        if largest is None or num>largest:
            second = largest
            largest = num
        if num != largest and (second is None or num > second):
            second = num
            
    return second

print(second_largest([1, 2, 3, 4, 5]))
print(second_largest([5, 5, 5]))
print(second_largest([10]))
print(second_largest([3, 1, 4, 4, 2]))
print(second_largest([-5, -2, -10]))
```

#### Write a function to group a list of words that are anagrams of each other.
Return a list of groups (the order of groups doesn't matter).

```python
from collections import defaultdict

def group_anagrams(words: list[str]) -> list[list[str]]:
    group = defaultdict(list)
    for word in words:
        key = "".join(sorted(word))
        group[key].append(word)
    return list(group.values())

print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
print(group_anagrams([""]))
print(group_anagrams(["abc"]))
```

#### Write a function to rotate a list to the right by k positions.

```python
def rotate(nums: list, k: int) -> list:
    if not nums:
        return nums
    k %= len(nums)
    return nums[-k:] + nums[:-k]

print(rotate([1,2,3,4,5], 2))
print(rotate([1,2,3], 4))
print(rotate([], 3))
```

#### Write a function to find the intersection of two lists.
Requirements: Return only unique elements. | The order of elements in the result does not matter.

```python
def intersection(list1:list, list2:list) -> list:
    set_list2 = set(list2)
    common_set = set()
    for i in list1:
        if i in set_list2:
            common_set.add(i)
    return list(common_set)

print(intersection([1,2,2,3], [2,2,4]))
print(intersection([1,2,3], [4,5,6]))
print(intersection([1,2,3], [3,2,1]))
```

#### Write a function to flatten a nested list of arbitrary depth.

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

print(flatten([1, [2, 3], [4, [5, 6]], 7]))
print(flatten([]))
print(flatten([1, [2], [[3]], [[[4]]]]))
```

Return the first element whose second occurrence appears first.
