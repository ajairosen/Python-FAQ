### Aggregation in OOP

```python
class BasicInfo:
    
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address
    
    def print_info(self):
        print(f"""
            Name: {self.name}\n
            Age: {self.age}\n
            Locality: {self.address.locality}\n
            Pin: {self.address.pin}""")
        

class Address:
    
    def __init__(self, locality, pin):
        self.locality = locality
        self.pin = pin
```

```python
address = Address("KK", 1234)

p1 = BasicInfo("Ajai", 27, address)
```

```python
p1.print_info()
```
