```python
class Person:
    def __init__(self, name, country):
        self.name = name
        self.country = country
        
    def greeting(self):
        if self.country == "india":
            print(f'Namaste {self.name}')
        else:
            print(f"Hello {self.name}")
            
p = Person("Ajai", "India")
p.greeting()
```

```python
## Attribute creation outside class
p.age = 27 
```
