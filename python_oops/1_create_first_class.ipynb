```python
class AtmMachine:
    
    def __init__(self):
        self.pin = 0
        self.balance = 0
        self.menu()
    
    def menu(self):
        user_input = input("""Select the below: 
                               Press 1 -> Create pin
                               Press 2 -> Change pin
                               Press 3 -> Check balance
                               Press 4 -> Withdraw balance
                               Press 5 or Anything -> Exit application""")
        if user_input == "1":
            #Create pin
            self.create_pin()

        elif user_input == "2":
            #Change pin
            self.change_pin()
        elif user_input == "3":
            #Check balance
            self.check_balance()
        elif user_input == "4":
            #Withdraw balance
            self.withdraw_balance()
        else:
            return
            
    def create_pin(self):
        self.pin = int(input('Enter the pin: '))
        
        self.balance = int(input("Enter user balance"))
        
        self.menu()
        
    def change_pin(self):
        old_pin = int(input('Enter old pin: '))
        if old_pin == self.pin:
            new_pin = int(input("Enter new pin: "))
            if new_pin == self.pin:
                print('Dont enter the old pin')
                # self.menu()
            else:
                self.pin = new_pin
                # self.menu()
        else:
            print("Please enter the correct pin")
        self.menu()
            
    def check_balance(self):
        pin = int(input('Enter your pin:'))
        if self.pin == pin:
            print(f"Your current balance is {self.balance}")
        else:
            print('Incorrect PIN entered')
        self.menu()
            
    def withdraw_balance(self):
        pin = int(input('Enter your pin: '))
        if self.pin == pin:
            withdraw_amount = int(input("Enter the amount you want to withdraw: "))
            if withdraw_amount <= self.balance:
                self.balance -= withdraw_amount
                print(f'{withdraw_amount} from your account. Current balance: {self.balance}')
            else:
                print("Insufficient balance")
        else:
            print("Wrong PIN")
        self.menu()
            
```
